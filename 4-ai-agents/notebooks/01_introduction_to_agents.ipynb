## ğŸ“š Key Concepts Explained

### What is an Autonomous Agent?

An **autonomous agent** is a system that can:
1. **Perceive** - Gather information about its environment
2. **Reason** - Think about what it perceives and decide what to do
3. **Act** - Take actions to achieve goals
4. **Learn** - Improve from the results of its actions

**Think of it like:** A human IT operator, but automated and always available.

### Agent vs. Traditional Automation

**Traditional Automation:**
- Follows fixed rules: "If service is down, restart it"
- No reasoning or adaptation
- Works well for predictable scenarios

**Autonomous Agent:**
- Uses AI to reason: "Service is down. Let me check why, then decide the best fix"
- Can adapt to new situations
- Learns from experience

**Example:**
- **Automation:** "If CPU > 90%, restart service"
- **Agent:** "CPU is high. Let me check if it's a temporary spike or sustained load. If sustained, I'll scale up resources. If spike, I'll investigate the cause first."

### Tools: What Agents Can Do

## ğŸ’¼ How This Applies to IT Operations

**Real-world impact:**
- **Reduce MTTR (Mean Time To Resolve):** Agents can detect and fix issues faster than humans
- **Scale operations:** One agent can monitor hundreds of services simultaneously
- **Learn from incidents:** Agents remember what worked and improve over time
- **Handle edge cases:** Agents can reason about situations they haven't seen before

## 1. Understanding the Simulated Environment

**What we're doing:** Creating a simulated IT environment to safely test agents.

**Why:** We need a safe place to experiment before using agents in real systems. Think of it like a flight simulator for pilots - it lets us practice without risk. No production systems will be harmed in the making of this notebook! ğŸ˜„

**What to expect:**
- We'll create a simulated environment with IT services (web-server, database, etc.)
- Each service has status, CPU usage, memory usage - just like real services
- The agent can interact with this environment safely - break things, fix things, learn things!


````python
# Create a simulated IT environment
# This simulates a real IT infrastructure with services

# Define the services in our environment
services = [
    "web-server",
    "database",
    "cache-service",
    "api-gateway",
    "monitoring-service"
]

# Create the environment
env = SimulatedEnvironment(initial_services=services)

print("âœ… Simulated environment created!")
print(f"ğŸ“Š Services in environment: {len(env.services)}")
print("\nğŸ” Current service status:")
for service_name in services:
    status = env.get_service_status(service_name)
    if status:
        icon = "âœ…" if status["status"] == "running" else "âš ï¸" if status["status"] == "degraded" else "âŒ"
        print(f"   {icon} {service_name}: {status['status']} (CPU: {status['cpu_usage']:.1f}%, Memory: {status['memory_usage']:.1f}%)")
````

## 2. Understanding Tools

**What we're doing:** Exploring the tools (actions) that agents can use.

**Why:** Tools are what give agents the ability to take actions. Without tools, agents can only think but not act - like having a brilliant mind but no hands! Tools are the agent's hands.

**What to expect:**
- We'll see what tools are available (check status, restart services, scale resources, etc.)
- Each tool has a name, description, and parameters - just like functions
- Tools are like functions - they take inputs and produce outputs


````python
# Create a tool registry
# This manages all the tools (actions) the agent can take

tool_registry = ToolRegistry(environment=env)

# List all available tools
print("ğŸ› ï¸  Available Tools:")
print("=" * 60)
tools = tool_registry.list_tools()
for i, tool in enumerate(tools, 1):
    print(f"\n{i}. {tool['name']}")
    print(f"   Description: {tool['description'][:100]}...")
    print()

print(f"\nâœ… Total tools available: {len(tools)}")
````

## 3. Testing Tools Manually

**What we're doing:** Manually executing tools to understand how they work.

**Why:** Before the agent uses tools automatically, let's see how they work manually. This helps us understand what the agent will be doing - think of it as learning to drive before letting the car drive itself!

**What to expect:**
- We'll call tools directly (like calling a function)
- We'll see the results of each action - success, failure, or something in between
- This demonstrates the "Act" part of the agent loop - where thinking becomes doing


````python
# Example 1: Check status of a specific service
print("ğŸ“Š Example 1: Checking service status")
print("-" * 60)
result = tool_registry.execute_tool("check_service_status", service_name="web-server")
print(result)
print()

# Example 2: Get status of all services
print("ğŸ“Š Example 2: Getting all services status")
print("-" * 60)
result = tool_registry.execute_tool("get_all_services")
print(result)
print()

# Example 3: Simulate a problem (for demonstration)
print("âš ï¸  Example 3: Simulating a service failure")
print("-" * 60)
env.simulate_failure("web-server")
result = tool_registry.execute_tool("check_service_status", service_name="web-server")
print(result)
print()

# Example 4: Fix the problem by restarting
print("ğŸ”§ Example 4: Restarting the failed service")
print("-" * 60)
result = tool_registry.execute_tool("restart_service", service_name="web-server")
print(result)
print()

# Verify it's fixed
print("âœ… Verifying the fix")
print("-" * 60)
result = tool_registry.execute_tool("check_service_status", service_name="web-server")
print(result)
````

## 4. Understanding Agent Reasoning

**What we're doing:** Seeing how an agent would reason about a problem.

**Why:** The key difference between automation and agents is **reasoning**. Agents think before acting - they don't just blindly follow rules. This is the "magic" that makes agents smart!

**What to expect:**
- We'll simulate what an agent would think (we'll show you its "thought process")
- The agent analyzes the situation (observes the problem)
- The agent decides which tools to use (plans the solution)
- The agent executes actions and evaluates results (acts and learns)

**The fun part:** In this notebook, we're showing the concept with simulated reasoning. In the next notebook, we'll build a real agent using LlamaStack that actually reasons using an LLM!


````python
# Simulate agent reasoning process
# This shows what an agent would think through

print("ğŸ¤– Agent Reasoning Simulation")
print("=" * 60)

# Reset environment for clean demonstration
env.reset()

# Simulate a problem: high CPU on database
env.simulate_degradation("database")

print("\nğŸ“‹ Step 1: OBSERVE - Agent gathers information")
print("-" * 60)
all_services = tool_registry.execute_tool("get_all_services")
print(all_services)

print("\nğŸ’­ Step 2: THINK - Agent analyzes the situation")
print("-" * 60)
print("""
Agent's reasoning:
- Database service shows 'degraded' status
- CPU usage is very high (85-95%)
- Memory usage is also high (80-90%)
- This indicates the service is overloaded
- Options:
  1. Restart the service (might help temporarily)
  2. Scale up the service (add more resources - better for sustained load)
- Since this is a sustained high load, scaling is the better option
""")

print("\nğŸ¯ Step 3: ACT - Agent decides to scale the service")
print("-" * 60)
result = tool_registry.execute_tool("scale_service", service_name="database", replicas=3)
print(result)

print("\nğŸ” Step 4: VERIFY - Agent checks if the action worked")
print("-" * 60)
status_after = tool_registry.execute_tool("check_service_status", service_name="database")
print(status_after)

print("\nğŸ“š Step 5: LEARN - Agent remembers this solution")
print("-" * 60)
print("""
Agent learns:
- Problem: High CPU/memory on database
- Solution: Scale to 3 replicas
- Result: CPU and memory reduced
- This pattern can be reused for similar problems
""")

print("\nâœ… Agent reasoning simulation complete!")
````
